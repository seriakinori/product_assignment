## BERTによるニュース分類（PyTorch） (<u>ライブラリを使用する</u>)

BERTとは、2018年にGoogleによって公開された「双方向Transomerによる汎用的言語表現モデル」です。

双方向Transformerによってテキストデータを事前学習することで、前後の文脈を考慮した分散表現を獲得しています。

BERTは下図のように、既存のタスク実行モデルに「くっつける」ものです。

この「くっつける」作業をFine-tuning、または転移学習と呼び、モデルの精度を向上させる役割を果たします。

BERTの構造は下図のようになっています。

ベクトルのシーケンス$(E_1 \cdot\cdot\cdot E_n)$を入力とし、ベクトルのシーケンス$(T_1 \cdot\cdot\cdot T_n)$が出力されるモデルです。

入力層と出力層の間に隠れ層`Trm(=Transformer)`を複数重ねた構造です。

bert.png

### Masked Language Model (MLM)
Masked Language ModelはBERTの学習方法の一つです。

ここでは文中の単語をマスクで隠し、予測することで学習しています。

マスクを予測するためにマスク部分の前後を理解する必要があり、これにより文章の前後の文脈を学習します。

````python
# BERTモデルとトークナイザをロードします
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForMaskedLM.from_pretrained("bert-base-uncased")

# 予測するテキストの用意。[MASK]部分を予測。
text = "I want to go to [MASK] on my vacation."
masked_ind = tokenizer.encode(text, add_special_tokens=True).index(103)
input = tokenizer.encode_plus(text, return_tensors="pt")
model.eval()
with torch.no_grad():
    output = model(**input)

# 予測された全単語のスコアを取得
scores = output[0][0, masked_ind]
top_5_ids = torch.topk(scores, 5).indices
num = 1
for id in top_5_ids:
    word = tokenizer.decode([id])
    print(f"{num}番目の単語 : {word}")
    print(f"文章 : I want to go to {word} on my vacation.")
    print("----------------------------------------------------------------")
    num += 1
````

### Next Sentence Prediction (NSP)
Next Sentence PredictionもまたBERTの事前学習方法の一つです。

ここでは2つの文章がある時、2番目の文章が1番目の文章に連続するかを予測し、

文章同士の関係性について学習することができます。

````python
model = BertForNextSentencePrediction.from_pretrained("bert-base-uncased")
# 予測するテキストを定義します
text = ("Python is my favorite language.", "I am learning deep learning.")
# 以下の場合連続した文章の可能性は0.26%
# text = ("Python is my favorite language.", "But our oceans are in danger of contamination.")

encoding = tokenizer.encode_plus(text[0], text[1], return_tensors="pt")
print(f"\nencodingの中身\n{encoding}\n")
model.eval()
with torch.no_grad():
    outputs = model(**encoding, labels=torch.LongTensor([1]))
logits = outputs.logits
probs = torch.nn.functional.softmax(logits, dim=1)
print(f"これらの文章が連続している確率は {probs[0][0]*100}%です")
````

### Tokenizerの用意

BERTはニューラルネットワークを用いた言語モデルが実装されていますが、一から実装するのは非常に手間がかかります。

そのため実際上は学習済みモデルを使用します。

今回は東北大学の研究チームによって作成されたBERTの日本語の事前学習モデル```cl-tohoku/bert-base-japanese-whole-word-masking```を用います。

この事前学習モデルはWikipediaの日本語記事のデータを用いて学習されています。

### 変数encodingの説明
#### input_ids
2つの文章をまとめてトークンにしています。input_ids内の数字はトークンのIDを指します。

#### token_type_ids
2つの文章を区別するためにあります。

上の変数encodingならtoken_type_idsは1~8番目が0なため、input_idsの先頭8個のトークンは前半の文章を示すことになります。

#### attention_mask
トークンに対してはパディングすることが可能です。

この時attention_maskでは普通の単語を指すトークンには1、パディング部分には0をつけます。

-------------------------------------------------------------------

* <font color="Red">問2. スライド11章で紹介されているBERTの関数を参考に、```bert-base-japanese-whole-word-masking```モデルのTokenizerを呼び出そう。ただし、日本語のTokenizerを呼び出す為、```BertTokenizer```の代わりに```BertJapaneseTokenizer```を使用しよう。</font><br>

````python
# トークナイザの用意
tokenizer = BertJapaneseTokenizer.from_pretrained(
    "cl-tohoku/bert-base-japanese-whole-word-masking"
)  ### 問2 ###

# データローダの為のデータセットを作成
n_label = 2
max_length = 40
category_list = category_list[:n_label]  # n_labelクラスだけ分類
dataset = []
for label, category in enumerate(tqdm(category_list)):
    for f in glob.glob(f"./data/livedoor/text/{category}/{category}*"):
        lines = open(f, encoding="utf-8_sig").read().splitlines()
        text = lines[2]  # タイトルのみ抽出
        encoding = tokenizer(
            text, max_length=max_length, padding="max_length", truncation=True
        )
        encoding["label"] = label
        encoding = {k: torch.tensor(v) for k, v in encoding.items()}
        dataset.append(encoding)
````

````python
# データの確認
print("データ数: ", len(dataset))
print(dataset[0])
````

````python
# データセットの分割
dataset_for_loader = dataset.copy()
random.seed(0)
random.shuffle(dataset_for_loader)
n = len(dataset_for_loader)
n_train = int(n * 0.6)
n_valid = int(n * 0.2)
dataset_train = dataset_for_loader[:n_train]
dataset_valid = dataset_for_loader[n_train : n_train + n_valid]
dataset_test = dataset_for_loader[n_train + n_valid :]

# データローダの作成
batch_size = 16

dataloader_train = DataLoader(dataset_train, batch_size=batch_size, shuffle=True)
dataloader_valid = DataLoader(dataset_valid, batch_size=batch_size, shuffle=False)
dataloader_test = DataLoader(dataset_test, batch_size=batch_size, shuffle=False)
print(len(dataset_train), len(dataset_valid), len(dataset_test))
print(len(dataloader_train), len(dataloader_valid), len(dataloader_test))
````

### 分類器の用意

BERTモデルを組み込んだ分類器を用意する。

````python
class BertClassifier(nn.Module):
    """BERTベースの分類器
    BERTモデルをベースにした分類器クラスです。事前学習されたBERTモデルの出力を線形層に通して分類を行います。

    Attributes:
        model_bert (BertModel): 事前学習されたBERTモデル。
        linear (torch.nn.Linear): 分類のための線形層。

    Examples:
        >>> bert_classifier = BertClassifier(n_label=2)
        >>> input_ids = torch.tensor([[1, 2, 3, 4], [5, 6, 7, 8]])
        >>> output = bert_classifier(input_ids)
        >>> print(output.shape)  # 例: torch.Size([2, 2])

    Note:
        - このクラスは、BERTの最終層の出力を利用して分類を行います。
        - `forward`メソッドでは、入力として単語IDの文章を受け取り、BERTモデルを通して得られた特徴から分類を行います。
    """

    def __init__(self):
        """BERTベースの分類器の初期化

        Note:
            - 事前学習されたBERTモデルと線形層を初期化します。
            - 線形層の重みとバイアスは正規分布で初期化されます。
        """
        super(BertClassifier, self).__init__()
        self.model_bert = model_bert
        self.linear = nn.Linear(
            in_features=768, out_features=n_label
        )  # in_features:BertModelの出力, output_features:分類ラベルの数
        nn.init.normal_(self.linear.weight, std=0.02)
        nn.init.normal_(self.linear.bias, 0)

    def forward(self, input_ids):
        """
        分類器の順伝播

        Args:
            input_ids (torch.Tensor): 単語IDの文章。サイズは [batch_size, max_length]。

        Returns:
            torch.Tensor: 分類結果。サイズは [batch_size, n_label]。

        Note:
            - 入力された単語IDの文章をBERTモデルで処理し、その出力を線形層で分類します。
            - BERTモデルの最初のトークンの出力を使用して分類を行います。
        """
        # input_ids: torch.Size([batch_size, max_length])の単語IDの文章

        output_bert = self.model_bert(input_ids)

        output_bert = output_bert[0]
        output_bert = output_bert[:, 0, :].view(-1, 768)  # [batch_size, hidden_size]に変換
        output = self.linear(output_bert)

        return output
````

### 学習，評価


事前学習モデルの最後のレイヤと、追加した新しい層の重みをファインチューニングによって調整します。

ファインチューニングを復習しておきましょう。

ファインチューニングは事前学習済みモデルの重みを初期値として、再学習で微調整する手法です。

今回は事前学習済みモデルの最後のレイヤの重みのみ微調整します。


fine-tuning.png

#### ファインチューニングと転移学習の違いと使い分け

ファインチューニングは計算量が多くなる一方、パラメータの更新が多いため精度が向上しやすくなります。

転移学習はファインチューニングよりも少ないデータセットで行うことが可能です。

モデルの設定をファインチューニング用に変更して学習させます。

この学習には10～20分ほどかかります。

````python
model = BertClassifier()

# ファインチューニングの設定
# 全ての勾配計算をFalseに
for param in model.parameters():
    param.requires_grad = False

# BertModelの最後のレイヤのみ勾配計算ありに変更
for param in model.model_bert.encoder.layer[-1].parameters():
    param.requires_grad = True

# BertClassifierのLinearレイヤを勾配計算ありに変更
for param in model.linear.parameters():
    param.requires_grad = True


optimizer = optim.Adam(
    [
        {"params": model.model_bert.encoder.layer[-1].parameters(), "lr": 5e-5},
        {"params": model.linear.parameters(), "lr": 1e-4},
    ]
)
criterion = nn.CrossEntropyLoss()


n_epochs = 5
for epoch in range(n_epochs):
    loss_train = 0
    loss_valid = 0
    accuracy_train = 0
    accuracy_valid = 0

    model.train()
    for batch in dataloader_train:
        x = batch["input_ids"]
        t = batch["label"]
        batch_size_iter = len(x)
        model.zero_grad()
        optimizer.zero_grad()

        output = model(x)
        loss = criterion(output, t)
        loss.backward()
        optimizer.step()
        pred = output.argmax(dim=1)

        loss_train += loss.item() * batch_size_iter
        accuracy_train += torch.sum(pred == t.data)

    model.eval()
    for batch in dataloader_valid:
        x = batch["input_ids"]
        t = batch["label"]
        batch_size_iter = len(x)

        output = model(x)
        loss = criterion(output, t)
        pred = output.argmax(dim=1)

        loss_valid += loss.item() * batch_size_iter
        accuracy_valid += torch.sum(pred == t.data)

    average_loss_train = loss_train / len(dataset_train)
    average_loss_valid = loss_valid / len(dataset_valid)
    total_accuracy_train = accuracy_train / len(dataset_train)
    total_accuracy_valid = accuracy_valid / len(dataset_valid)
    print(
        "| epoch %d | train loss %.2f, train accuracy %.2f | valid loss %.2f, valid accuracy %.2f"
        % (
            epoch + 1,
            average_loss_train,
            total_accuracy_train,
            average_loss_valid,
            total_accuracy_valid,
        )
    )
````

#### ドメインシフト
ドメインシフトとは、訓練データとテストデータの分布が異なることを指します。

ドメインシフトの課題としてモデルの汎化性能が悪化し、テストデータでのパフォーマンスが低下することが挙げられます。

#### ドメイン適応
ドメイン適応とは、ドメインシフトを克服するための技術です。

ドメイン適応はタスクが同じでも学習データとテストデータの分布が異なる時に有効で、テストデータのラベルの有無によってドメイン適応の手法が異なります。

### 転移学習

転移学習はあるタスクで学習したモデルを別のタスクに応用する際に用いられます。

転移学習は主に出力層のパラメータを変更します。


````python
model = BertClassifier()

# 転移学習の設定
# 全ての勾配計算をTrueに
for param in model.parameters():
    param.requires_grad = True

optimizer = optim.Adam(model.parameters(), lr=1e-4)
criterion = nn.CrossEntropyLoss()

n_epochs = 5
for epoch in range(n_epochs):
    loss_train = 0
    loss_valid = 0
    accuracy_train = 0
    accuracy_valid = 0

    model.train()
    for batch in dataloader_train:
        x = batch["input_ids"]
        t = batch["label"]
        batch_size_iter = len(x)
        model.zero_grad()
        optimizer.zero_grad()

        output = model(x)
        loss = criterion(output, t)
        loss.backward()
        optimizer.step()
        pred = output.argmax(dim=1)

        loss_train += loss.item() * batch_size_iter
        accuracy_train += torch.sum(pred == t.data)

    model.eval()
    for batch in dataloader_valid:
        x = batch["input_ids"]
        t = batch["label"]
        batch_size_iter = len(x)

        output = model(x)
        loss = criterion(output, t)
        pred = output.argmax(dim=1)

        loss_valid += loss.item() * batch_size_iter
        accuracy_valid += torch.sum(pred == t.data)

    average_loss_train = loss_train / len(dataset_train)
    average_loss_valid = loss_valid / len(dataset_valid)
    total_accuracy_train = accuracy_train / len(dataset_train)
    total_accuracy_valid = accuracy_valid / len(dataset_valid)
    print(
        "| epoch %d | train loss %.2f, train accuracy %.2f | valid loss %.2f, valid accuracy %.2f"
        % (
            epoch + 1,
            average_loss_train,
            total_accuracy_train,
            average_loss_valid,
            total_accuracy_valid,
        )
    )
````
